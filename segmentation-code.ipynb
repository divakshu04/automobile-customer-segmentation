BALANCING THE DATASET

```python
preprocessed_data['Segmentation'].value_counts()
```

```python
equal_count = 1858
three_counter = 0
two_counter = 0
zero_counter = 0
indices_to_remove = []
for i in range(preprocessed_data['Segmentation'].shape[0]):
    if preprocessed_data['Segmentation'].values[i] == 3:
        three_counter += 1
        if three_counter > equal_count:
            indices_to_remove.append(i)
    elif preprocessed_data['Segmentation'].values[i] == 2:
        two_counter += 1
        if two_counter > equal_count:
            indices_to_remove.append(i)
    elif preprocessed_data['Segmentation'].values[i] == 0:
        zero_counter += 1
        if zero_counter > equal_count:
            indices_to_remove.append(i)
preprocessed_data = preprocessed_data.drop(indices_to_remove)
```

```python
preprocessed_data['Segmentation'].value_counts()
```

INPUTS AND TARGETS

```python
preprocessed_data.shape
```

```python
unscaled_inputs = preprocessed_data.iloc[:, :-1]
unscaled_targets = preprocessed_data.iloc[:, -1]
```

SCALING DATASET

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
columns_to_scale = ['Work_Experience','Family_Size']
preprocessed_data[columns_to_scale] = scaler.fit_transform(preprocessed_data[columns_to_scale])
preprocessed_data.head()
```

```python
df_combined = pd.concat([unscaled_inputs, preprocessed_data['Segmentation']], axis=1)
df_shuffled = df_combined.sample(frac=1, random_state=365)
shuffled_inputs = df_shuffled.drop(['Segmentation'], axis=1)
shuffled_targets = df_shuffled['Segmentation']
```

```python
shuffled_inputs.shape
```

SPLITTING DATASET

```python
samples_count = shuffled_inputs.shape[0]
train_sample_count = int(0.8 * samples_count)
validation_sample_count = int(0.1 * samples_count)
test_sample_count = samples_count - train_sample_count - validation_sample_count

train_inputs = shuffled_inputs[: train_sample_count]
train_targets = shuffled_targets[: train_sample_count]

validation_inputs = shuffled_inputs[train_sample_count : train_sample_count + validation_sample_count]
validation_targets = shuffled_targets[train_sample_count : train_sample_count + validation_sample_count]

test_inputs = shuffled_inputs[train_sample_count + validation_sample_count :]
test_targets = shuffled_targets[train_sample_count + validation_sample_count :]

print(np.sum(train_targets) / train_sample_count)
print(np.sum(validation_targets) / validation_sample_count)
print(np.sum(test_targets) / test_sample_count)
```

NEURAL NETWORK MODEL

```python
np.savez('segmentation_data_train', inputs=train_inputs, targets=train_targets)
np.savez('segmentation_data_validation', inputs=validation_inputs, targets=validation_targets)
np.savez('segmentation_data_test', inputs=test_inputs, targets=test_targets)
```

```python
npz = np.load('segmentation_data_train.npz')
processed_train_inputs = npz['inputs'].astype(np.float32)
processed_train_targets = npz['targets'].astype(np.int64)

npz = np.load('segmentation_data_validation.npz')
processed_validation_inputs = npz['inputs'].astype(np.float32)
processed_validation_targets = npz['targets'].astype(np.int64)

npz = np.load('segmentation_data_test.npz')
processed_test_inputs = npz['inputs'].astype(np.float32)
processed_test_targets = npz['targets'].astype(np.int64)
```

```python
processed_train_inputs.shape
```

```python
import tensorflow as tf
input_size = 23
output_size = 4
hidden_layer_size = 100
model = tf.keras.Sequential([
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(output_size, activation='softmax')
])
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(optimizer=Adam(learning_rate = 0.002), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
batchsize = 100
max_epochs = 100
early_stopping = tf.keras.callbacks.EarlyStopping(patience=2)
model.fit(processed_train_inputs,
          processed_train_targets,
          batch_size = batchsize,
          epochs = max_epochs,
          callbacks = [early_stopping],
          validation_data = (processed_validation_inputs, processed_validation_targets),
          verbose = 2
          )
```

TESTING THE MODEL

```python
test_loss, test_accuracy = model.evaluate(processed_test_inputs, processed_test_targets)
```

```python
print('Test loss : {0:.2f}   Test accuracy : {1:.2f}%'.format(test_loss, test_accuracy*100))
```
